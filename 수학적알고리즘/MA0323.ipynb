```python
import pandas as pd
w = pd.read_csv("ch5-1.csv")
```

```python
w.head()
```

```python
w.info()
```

```python
w_n = w.iloc[:,1:5]
w_n.head()
```

```python
w_n.describe()
```

```python
w_cor = w_n.corr(method = "pearson")
w_cor
```

```python
w_cor.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(w_n ,kind='reg')
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(w_cor, annot = True, cmap = 'Blues')
plt.show()
```

병아리알의 무게, 사료 >> 닭의 성장 예측 모델

종란의 무게와 매일 사료를 얼만큼 먹었냐 따져서 닭이 도축할때 무게가 얼마가 될지 예측하는 모델

일단 로그 함수 형태로 나타날것으로 예측

"모델" 이 핵심!!

닭을 도축할때 == 70일

알의 무게를 보고 평균 닭의 무게를 예측한뒤 실제 와의 오차가 어떻게 되는지 조사

1.참값이 있으면 그것을 관측하는 것은 다양한 요인으로 참값이 아닐수도있다

참값을 기준으로 관측값은 정규분포가 그려진다

2.관측값을 중심으로 정규분포를 그린다.(히스토그램으로 그리면 정규분포를 그린다)
