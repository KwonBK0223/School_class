Spline Interpoloation으로 시각화

Spline 보간법 : 전체 구간을 소구간별로 나누어 저차수의 다항식으로 매끄러운 함수를 구하는 방법

# PM10

```python
k1.plot(kind='line',c='r',marker='.',mec='b')
plt.xticks(range(24))
plt.yticks(range(10,35))
plt.grid(True)
plt.show()

x=np.arange(0,24,1)
y=[]
for i in range(24):
    y.append(t1.iloc[i,0])
y=np.array(y)

f_quadr = interpolate.interp1d(x, y, kind='quadratic')

xnew=[]
for i in np.arange(0,23,0.00001):
    xnew.append(i)

y_new_quadr = f_quadr(xnew)

plt.plot(x, y, "o", xnew, y_new_quadr)
plt.xticks(range(24))
plt.yticks(range(10,35))
plt.grid(True)
plt.show()
```
