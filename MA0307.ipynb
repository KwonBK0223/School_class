```python
import numpy as np
x=np.arange(0,24,1)
y=[]
for i in range(24):
    y.append(a1.iloc[i,0])
y=np.array(y)
```

```python
from scipy import interpolate
import matplotlib.pyplot as plt

f_quadr = interpolate.interp1d(x, y, kind='quadratic')

xnew=[]
for i in np.arange(0,23,0.0001):
    xnew.append(i)

y_new_quadr = f_quadr(xnew)

plt.plot(x, y, "o", xnew, y_new_quadr)
plt.xticks(range(24))
plt.yticks(range(10,35))
plt.grid(True)
plt.show()
```
